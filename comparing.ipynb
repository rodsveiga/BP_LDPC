```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = '/home/rodsveiga/Dropbox/DOC/van_error_codes/models'

file_VANs = path + '/p_versus_overlap_N_250_M_1000_K_5_p_prior_0100_steps_1000_code_0'
file_BP = 'p_versus_overlap_BP_N_250_M_1000_p_prior_0100_messages_5_it_8.txt'
```
